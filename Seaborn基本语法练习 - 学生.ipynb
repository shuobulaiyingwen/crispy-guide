# 绘制散点图及其参数设置

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# 使用默认主题
# sns.set_theme()
sns.set()
```

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

```python
# 散点图
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# 散点图
                                            # 默认散点图scatter
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```python
# 通过色相 hue 区分不同的类别
sns.relplot(data=tips,x='total_bill',y='tip',
            hue='smoker',kind='scatter')
```

```python
# 通过色相hue、形状style两钟编码进行区分
sns.relplot(data=tips,x='total_bill',y='tip',
            hue='smoker',style='smoker',kind='scatter')
```

```python
# hue为连续数值
sns.relplot(data=tips,x='total_bill',y='tip',
            hue='size',kind='scatter')
```

```python
# 利用点的大小进行区分
sns.relplot(data=tips,x='total_bill',y='tip',
            hue='smoker',kind='scatter')
```

```python
# 规定点大小的变化区间
sns.relplot(data=tips,x='total_bill',y='tip',
            size='size',sizes=(1,200),kind='scatter')
```

```python
# 通过绘制子图区分不同的用餐时间，有利于对比分析
sns.relplot(data=tips,x='total_bill',y='tip',
            hue='smoker',col='time',kind='scatter')
```

# 折线图

```python
# 加载航空数据集
flights = pd.read_csv('flights.csv')
flights.head()
```

```python
# 绘制折线图(复杂数据会显示置信区间)
sns.relplot(data=flights,x='year',y='passengers',kind='line')
```

```python
# 不显示置信区间，errorbar=None
sns.relplot(data=flights,x='year',y='passengers',kind='line',errorbar=None)
```

```python
# 历年每个月乘客数量变化
sns.relplot(data=flights,x='year',y='passengers',
            hue='month',kind='line')
```

```python
# 缩略组图
sns.relplot(data=flights,x='month',y='passengers',
            col='year',col_wrap=4,linewidth=2.5,height=3,kind='line')
plt.xticks([0,3,6,9])
```

# 柱形图

```python
#柱状图
sns.countplot(data=tips,x='day')
```

```python
# 分组柱形图
sns.countplot(data=tips,x='day',hue='sex')
```

```python
# 分组条形图
sns.countplot(data=tips,y='day',hue='sex')
# 添加标题
plt.rcParams['font.sans-serif']=['Microsoft YaHei']
plt.title("男女消费总量在不同时间段的统计对比图")
```
